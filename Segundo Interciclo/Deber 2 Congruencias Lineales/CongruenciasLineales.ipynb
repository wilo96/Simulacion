```python
import pandas as pd
```

```python
a=int(input("Ingrese a que es numero primo: "))
b=int(input("Ingrese b que es numero primo: "))
m=int(input("Ingrese m que es numero primo: "))
x0=int(input("Ingrese Semilla: "))
itera=int(input("Ingrese numero de iteraciones: "))
i=0
Xn=0
Un=0
LXn=[]
LUn=[]
Liter=[]
Xn=x0
Liter.append(i)
i=i+1
LXn.append(Xn)
LUn.append(Un)
while(i <= itera):
    Liter.append(i)
    Xn= (a * Xn + b) % m
    LXn.append(Xn)
    Un=Xn/m
    LUn.append(Un)
    i=i+1
    
data={'Iteracion':Liter,'Xn':LXn,'Un':LUn}
df=pd.DataFrame(data)
df
```
