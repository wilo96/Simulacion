```python
import pandas as pd
```

```python
x0 = 9850
digitos=4
iteracion=0
Xn=x0
Ni=10
LXn=[]
LXc=[]
Li=[]
LL=[]
LUi=[]
LRn=[]
while(iteracion<=Ni):
    Xc=int(Xn) * int(Xn)
    Xcs=str(Xc)
    while(len(Xcs) < 8):
        Xcs="0"+Xcs

    Longitud=len(Xcs)
    Ui=Xcs[int((Longitud/2)-2):int((Longitud/2)+2)]
    Nn="1"
    while(len(Nn)<(digitos+1)):
        Nn=Nn+"0"
    Rn=int(Ui)/int(Nn)    
    #print("IteraciÃ³n: ",iteracion)
    Li.append(iteracion)
    #print("Xn: ",Xn)
    LXn.append(Xn)
    #print("Xn * Xn: ",Xcs)
    LXc.append(Xcs)
    #print("Longitud: ",Longitud)
    LL.append(Longitud)
    #print("Ui: ",Ui)
    LUi.append(Ui)
    #print(Nn)
    #print("Randomico: ",Rn)
    LRn.append(Rn)
    Xn=Ui
    iteracion=iteracion+1
    
data={'Iteracion':Li,'Xn':LXn,'Xn*Xn':LXc,'Longitud':LL,'Ui':LUi,'Rn':LRn}
df=pd.DataFrame(data)
print("Semilla: ",x0)
print("Numero de Iteraciones: ",Ni)
print("Digitos: ",digitos)
df
```
